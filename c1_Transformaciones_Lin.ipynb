```python
A = np.array([ [-1,3], [2,-2] ])
print('A = \n', A)
```

```python
v = np.array([[2],[1]])
print('v = \n',v)
```

```python
vf=v.flatten() # <--- es análogo a la transpuesta 'v.T'
print('vf = \n',vf)
```

```python
# Vector v tranformado:
w = np.dot(A,v)
print('A*v = \n', w)
```

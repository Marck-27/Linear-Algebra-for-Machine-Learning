# Eigenvectores y eigenvalores:

 Si $A$ es una matriz de $n \times n$

$\vec{v}$ y $\lambda$ son eigenvector y eigenvalor de $A$, respectivamente, si:

$A \vec{v} = \lambda \vec{v}$.

```python
# Matriz:
X = np.array([ [3,2],[4,1] ])
print('X = \n',X)
```

```python
# Obtenemos los Eigenvalores y Eigenvectores:
eigenval_X, eigenvec_X=np.linalg.eig(X)

print('Eigenvalores: ',eigenval_X)
print('\nEigenvectores (cada columna): \n',eigenvec_X)
```

```python
# Extraemos los eigenvalores:
lam1=eigenval_X[0]
lam2=eigenval_X[1]

# Extraemos los eigenvectores:
v1=eigenvec_X[:,0]
v2=eigenvec_X[:,1]
```

```python
# Comprobación:
Xv1 = np.dot(X,v1) # <-- A*v1
w1 = lam1*v1 # <-- l1*v1

Xv2 = np.dot(X,v2) # <-- A*v2
w2 = lam2*v2 # <-- l2*v2



print('X*v1 = \n',Xv1)
print('\nl1*v1 = \n',w1)

print('\n\nX*v2 = \n',Xv2)
print('\nl2*v2 = \n',w2)
```

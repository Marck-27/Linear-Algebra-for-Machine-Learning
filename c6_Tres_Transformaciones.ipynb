## Obtenemos los factores de SVD (Singular Values Decomposition):

```python
U, D, V = np.linalg.svd(A)
D =np.diag(D)

# Cada matriz represnta una serie de transformaciones que definene la matriz original:
# U <-- Primer rotación de la base
# D <-- Escalamiento (Valores singulaes)
# V <-- Segunda rotación de la base

print('U = \n',U)
print('\nD = \n',D)
print('\nV = \n',V)
```

```python
graficarMatriz2(I)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.title('Circulo Unitario')
plt.show()

graficarMatriz2(V)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.title('Primer rotación por V')
plt.show()

DV=np.dot(D,V) # <-- aplicamos D a V
graficarMatriz2(DV)
plt.xlim(-9,9)
plt.ylim(-9,9)
plt.title('Escalamiento por D')
plt.show()

UDV=np.dot(U,DV)# <-- aplicamos U a DV
graficarMatriz2(UDV)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.title('Segunda rotación por U')
plt.show()

graficarMatriz2(A)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.title('Comprobación A = UDV')
plt.show()
```
